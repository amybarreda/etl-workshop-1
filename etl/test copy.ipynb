```python
import pandas as pd

df = pd.read_csv('C:/Users/amygi/Documents/IDEIA/2026_01/ETL/Workshops/first/data/raw/candidates.csv',
                 sep=';')

print(df.head())
```

```python
# comprobamos si en el DF ya cargado hay correos repetidos
# (el csv ya se leyó en la celda anterior y está en `df`)

# filas donde el campo Email aparece más de una vez
dup_emails = df[df['Email'].duplicated(keep=False)]

print(f"filas con correo duplicado: {len(dup_emails)}")
if not dup_emails.empty:
    display(dup_emails.sort_values('Email'))
else:
    print("no se detectaron correos repetidos")

# opcional: lista de direcciones que se repiten
repetidas = df['Email'][df['Email'].duplicated()].unique()
print("direcciones que aparecen más de una vez:", repetidas)
```
